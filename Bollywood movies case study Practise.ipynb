```python
df['writers'].fillna('NA',inplace=True)
df['genre'].fillna('NA',inplace=True)
df['releaseDate'].fillna('NA',inplace=True)
df['releaseDate'].fillna('00/00/0000', inplace=True)
df['actors'].fillna('NA', inplace=True)
df['directors'].fillna('NA',inplace=True)
df['sequel'].fillna(0,inplace=True)
```

```python
df.head(10)
```

```python
df.rename(columns={'title':'Title','releaseYear':'Release Year','releaseDate':'Release Date',
                'genre':'Genre',   'writers':'writers', 'actors': 'Actors','directors':'Directors','Sequel':'Sequel',
                   'HitFlop':'Hit-Flop'}, inplace=True)
```
